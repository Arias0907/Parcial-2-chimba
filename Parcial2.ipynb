#**Punto 1**

En este primer punto realice dos mascaras las cuales eran en forma de una circunferencia, donde toco tener unos diferentes radios los cuales tenian una pequeña diferemcia uno a otro para que así el filtrado pareciera una dona, el principio le dije que solo leyera lo que estaba dentro del anillo de la dona, y para la segunda parte le dije que leyera todo menos el anillo de la dona. 

```
img = cv2.imread('flores.jpg', 0)
img_flotante = np.float32(img)
dft = cv2.dft(img_flotante, flags = cv2.DFT_COMPLEX_OUTPUT)
dft_shift=np.fft.fftshift(dft)
magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0], dft_shift[:,:,1]))
fils, cols = img.shape
centfils, centcols = fils/2, cols/2

#creemos unas mascara con un centro cuadrado de 1 y el resto zeros
mask = np.zeros((fils, cols, 2), np.uint8)
r = 30
r2=20
center = [centfils, centcols]
x, y = np.ogrid[:fils, :cols]

mask_area= (x - center[0])**2 + (y - center[1])**2 <= r*r

mask_areas= (x - center[0])**2 + (y - center[1])**2 <= r2*r2

mask[mask_area]=1;
mask[mask_areas]=0;

##Aplicquemos la mascara y luego obtenemos la inversa de DFT
fshift = dft_shift*mask
fshift_mask_mag = 2000* np.log(cv2.magnitude(fshift[:,:,0], fshift[:,:,1]))

f_shift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_shift)
img_back = cv2.magnitude(img_back[:,:,1], img_back[:,:,0])

displayGray4(img, img_back, magnitude_spectrum, fshift_mask_mag)

#creemos unas mascara con un centro cuadrado de 1 y el resto zeros
mask2 = np.ones((fils, cols, 2), np.uint8)
r3 = 80
r4 = 70
center = [centfils, centcols]
x1, y1 = np.ogrid[:fils, :cols]

mask3 = np.ones((fils, cols, 2), np.uint8)

mask_areaN = (x1 - center[0])**2 + (y1 - center[1])**2 <= r3*r3
mask_areasN = (x1 - center[0])**2 + (y1 - center[1])**2 <= r4*r4

mask2[mask_areaN]=0
mask2[mask_areasN]=1
   
##Aplicquemos la mascara y luego obtenemos la inversa de DFT
fshift1 = dft_shift*mask2
fshift_mask_mag1 = 2000* np.log(cv2.magnitude(fshift1[:,:,0], fshift1[:,:,1]))


f_shift = np.fft.ifftshift(fshift1)
img_back1 = cv2.idft(f_shift)
img_back1= cv2.magnitude(img_back1[:,:,1], img_back1[:,:,0])

displayGray4(img, img_back1, magnitude_spectrum, fshift_mask_mag1)
```

#**Punto 2**

En este punto utilice al principio un median blur, para quitarle el ruido del principio. despues de eso realice dos laplacianos en los que resalte unos puntos especificos para cada uno de los laplacianos, y para resaltar las lineas utilice 2 sharpened que recibian a los laplacianos y esos dos sharpened los sume para que las lineas quedaran mas resaltadas.  

```
img = cv2.imread("saltlgle.png",0)

median = cv2.medianBlur(img, 5)


kernel_laplaciano2 = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])

sharpened2 = cv2.filter2D(median, -1, kernel_laplaciano2)


kernel_laplaciano3 = np.array([[1, 1, 1], [1, -8, 1], [1, 1, 1]])

sharpened3 = cv2.filter2D(median, -1, kernel_laplaciano3)

sharpened4 = sharpened2 + sharpened3

displayGray2(img, sharpened4)
```
